```python
import numpy as np
import pandas as pd
from numpy.random import choice
import matplotlib.pyplot as plt
```

## Question 1

```python
n = 500
```

```python
def TS(n, last_slice):
    prop = np.array([])
    num_arms = 2
    for i in np.arange(1000):
        bandit_selected = np.array([])
        rewards = [0]*num_arms
        penalties = [0]*num_arms
        for t in np.arange(1, n + 1):
            prob_array = [(t - 1/2)/n, 1 - (t - 1/2)/n]
            bandit = 0
            beta_max = 0
            for i in np.arange(0, num_arms):
                beta_curr = np.random.beta(rewards[i] + 1, penalties[i]+ 1)
                if beta_curr > beta_max:
                    beta_max = beta_curr
                    bandit = i
            bandit_selected = np.append(bandit_selected, bandit)
            reward = np.random.binomial(1, prob_array[bandit])
            if reward == 1:
                rewards[bandit] += 1
            else:
                penalties[bandit] += 1

        last_piece = bandit_selected[n-last_slice:n] + 1
        prop = np.append(prop, np.sum([1 if i == 1 else 0 for i in last_piece])/last_slice)
    return prop
```

```python
plt.title('Proportion of Arm = 1 in the last 25 samples for 1000 draws with n = 500')
plt.xlabel('proportion of arm = 1')
plt.ylabel('counts')
plt.hist(TS(500, 25));
```
